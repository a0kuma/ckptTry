```python
import torch
import torch.nn as nn
import torch.utils.checkpoint as checkpoint
```

```python
torch.cuda.memory._record_memory_history()
```

```python
device = "cuda"
```

```python
model = nn.Sequential(
    nn.Linear(100, 4096),
    nn.Linear(4096, 4096),
    nn.Linear(4096, 10)
).to(device)
```

```python
# Dummy data
batch_size = 32
x = torch.randn(batch_size, 100, device=device)
y = torch.randint(0, 10, (batch_size,), device=device)
```

```python
y_prime = checkpoint.checkpoint_sequential(
    model,
    segments=3,
    input=x,
    use_reentrant=False
)
```

```python
crossEntropyLoss_func_obj=nn.CrossEntropyLoss()
loss=crossEntropyLoss_func_obj(y_prime,y)
```

```python
loss.backward()
```

```python
torch.cuda.memory._dump_snapshot('abc.pickle')
```
